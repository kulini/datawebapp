```python
# Import libraries 

import pandas as pd
```

```python
# Load datasets 

vehicles = pd.read_csv('/Users/esdros/Desktop/Sprint_4_Project/vehicles_us.csv')
```

```python
# Previewing the dataset 

vehicles.info()

vehicles.head()
```

```python

# Checking for duplicates

vehicles.duplicated().sum()
```

```python

# Checking for missing values 

vehicles.isna().sum()
```

```python
# Removing missing values

vehicles.isna().sum()

vehicles = vehicles.dropna()
```

```python
# Checking for missing values again

vehicles.isna().sum()
```

```python
# Histogram of the condition distribution
import plotly.express as px

fig = px.histogram(vehicles, x='condition')
fig.update_layout(title='Count of Cars by Condition', xaxis_title='Condition', yaxis_title='Count')
fig.show()
```

```python
# Histogram of price distribution
# I had to use matplotlib because pyplot.express was not working
import matplotlib.pyplot as plt

plt.hist(vehicles['price'], bins=50)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
#Scatterplot comparing the odometer reading and model year
plt.scatter(vehicles['model_year'], vehicles['odometer'])
plt.title('Odometer vs Model Year')
plt.xlabel('Model Year')
plt.ylabel('Odometer Reading')
plt.show()
```

```python
fig_scatter = px.scatter(vehicles, x='model_year', y='price')
fig_scatter.update_layout(title='Model Year vs Price', xaxis_title='Model Year', yaxis_title='Price')
fig_scatter.show()
```
